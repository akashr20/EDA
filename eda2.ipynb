### Data cleaning

```python
df = pd.read_csv(r"E:\Datascience\Data set\mental_health_diagnosis_treatment_.csv")
df.columns = [col.replace(" ", "_") for col in df.columns]
df.dropna(inplace=True)
numeric_column = 'numeric_column'
if numeric_column in df.columns:
    df[numeric_column].fillna(df[numeric_column].mean(), inplace=True)
categorical_column = 'categorical_column'
if categorical_column in df.columns:
    df[categorical_column].fillna('Unknown', inplace=True)
df.drop_duplicates(inplace=True)
print(df.shape)
df.isna().sum()
df.info()
```

### checking for duplicates

```python
print(df.duplicated().sum())
```

### statistical analysis

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
stat_an=numeric_data.describe()
print(stat_an)
print(df.head(20))
print(df.tail())
plt.figure(figsize=(15, 10))
sns.boxplot(data=numeric_data)
plt.title('Boxplot of Numeric Data')
plt.xticks(rotation=90)  
plt.show()
```

### performing correaltion

```python
corelation=numeric_data.corr()
print(corelation)
plt.figure(figsize=(12,8))
sns.heatmap(corelation,cmap='spring',annot=True,fmt='.2f',linewidths=0.5)
plt.title("corelation")
plt.show()
```

### performing covarience

```python
covarience=numeric_data.cov()
print(covarience)
plt.figure(figsize=(12,8))
sns.heatmap(covarience,cmap='rainbow',annot=True,fmt='.2f',linewidths=0.5)
plt.title("covarience")
plt.show()
```

### some questions regarding the dataset

### 1.What is the distribution of ages among patients?


```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title('Distribution of Ages Among Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(True)
plt.show()
```

### 2.What is the gender distribution of the patients?

```python
gender_distribution = df['Gender'].value_counts()
print(gender_distribution)
plt.figure(figsize=(8, 6))
gender_distribution.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Gender Distribution of Patients')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

### 3.What are the most common mental health diagnoses in the dataset?

```python
diagnosis_distribution = df['Diagnosis'].value_counts()
print(diagnosis_distribution)
plt.figure(figsize=(10, 7))
plt.pie(diagnosis_distribution, labels=diagnosis_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Most Common Mental Health Diagnoses')
plt.axis('equal') 
plt.show()
```

### 4.How does symptom severity vary across different diagnoses?

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='Diagnosis', y='Symptom_Severity_(1-10)', data=df)
plt.title('Symptom Severity Across Different Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Symptom_Severity_(1-10)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

### 5.What is the relationship between mood scores and sleep quality?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Mood_Score_(1-10)', y='Sleep_Quality_(1-10)', data=df)
plt.title('Relationship Between Mood Scores and Sleep Quality')
plt.xlabel('Mood Score (1-10)')
plt.ylabel('Sleep Quality (1-10)')
plt.grid(True)
plt.show()
```

### 6.How does physical activity (hours per week) correlate with symptom severity?

```python
subset_data = df[['Physical_Activity_(hrs/week)','Symptom_Severity_(1-10)']]

# Calculate the correlation matrix
correlation_matrix = subset_data.corr()
print(correlation_matrix)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap Between Physical Activity and Symptom Severity')
plt.show()

```

### 7.What are the most commonly prescribed medications for each diagnosis?

```python
diagnosis_col = 'Diagnosis'
medication_col = 'Medication'


medication_counts = df.groupby(diagnosis_col)[medication_col].value_counts().reset_index(name='Count')
most_common_medications = medication_counts.loc[medication_counts.groupby(diagnosis_col)['Count'].idxmax()]
print(most_common_medications)
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y=diagnosis_col, hue=medication_col, data=most_common_medications, dodge=False)
plt.title('Most Commonly Prescribed Medications for Each Diagnosis')
plt.xlabel('Count')
plt.ylabel('Diagnosis')
plt.legend(title='Medication')
plt.show()
```

### 8.Which therapy types are most frequently used in the dataset?

```python
therapy_type_col = 'Therapy_Type'

# Count the occurrences of each therapy type
therapy_counts = df[therapy_type_col].value_counts().reset_index()
therapy_counts.columns = [therapy_type_col, 'Count']

# Print the therapy counts
print(therapy_counts)

# Plot a bar graph
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y=therapy_type_col, data=therapy_counts)
plt.title('Most Frequently Used Therapy Types')
plt.xlabel('Count')
plt.ylabel('Therapy Type')
plt.show()
```

### 9.What is the average duration of treatment for different diagnoses?

```python
diagnosis_col = 'Diagnosis'
treatment_duration_col = 'Treatment_Duration_(weeks)'

# Calculate the average duration of treatment for different diagnoses
avg_treatment_duration = df.groupby(diagnosis_col)[treatment_duration_col].mean().reset_index()
avg_treatment_duration.columns = [diagnosis_col, 'Average Duration (weeks)']

# Print the average treatment duration
print(avg_treatment_duration)

# Plot a bar graph
plt.figure(figsize=(12, 8))
sns.barplot(x='Average Duration (weeks)', y=diagnosis_col, data=avg_treatment_duration)
plt.title('Average Duration of Treatment for Different Diagnoses')
plt.xlabel('Average Duration (weeks)')
plt.ylabel('Diagnosis')
plt.show()
```

### 10.How do stress levels vary across different diagnoses?

```python
diagnosis_col = 'Diagnosis'
stress_level_col = 'Stress_Level_(1-10)'

# Calculate the average stress level for different diagnoses
avg_stress_levels = df.groupby(diagnosis_col)[stress_level_col].mean().reset_index()
avg_stress_levels.columns = [diagnosis_col, 'Average Stress Level']

# Print the average stress levels
print(avg_stress_levels)

# Plot a bar graph
plt.figure(figsize=(12, 8))
sns.barplot(x='Average Stress Level', y=diagnosis_col, data=avg_stress_levels)
plt.title('Average Stress Levels for Different Diagnoses')
plt.xlabel('Average Stress Level')
plt.ylabel('Diagnosis')
plt.show()
```

### 11.What are the most common outcomes of treatments?

```python
outcome_col = 'Outcome'

# Count the occurrences of each treatment outcome
outcome_counts = df[outcome_col].value_counts().reset_index()
outcome_counts.columns = [outcome_col, 'Count']

# Print the treatment outcome counts
print(outcome_counts)

# Plot a bar graph
plt.figure(figsize=(12, 8))
sns.barplot(x='Count', y=outcome_col, data=outcome_counts)
plt.title('Most Common Treatment Outcomes')
plt.xlabel('Count')
plt.ylabel('Treatment Outcome')
plt.show()
```

### 12. How does treatment progress (scores) change over time?


```python
weekly_progress = (
    df.groupby('Treatment_Duration_(weeks)')['Treatment_Progress_(1-10)']
    .mean()
    .reset_index()
    .rename(columns={'Treatment_Duration_(weeks)': 'Weeks', 'Treatment_Progress_(1-10)': 'Avg Progress'})
)
print(weekly_progress)
# Plotting the treatment progress trend
plt.figure(figsize=(10, 6))
plt.plot(weekly_progress['Weeks'], weekly_progress['Avg Progress'], marker='o', color='b')
plt.title('Average Treatment Progress Over Time', fontsize=16)
plt.xlabel('Weeks of Treatment', fontsize=14)
plt.ylabel('Average Progress Score', fontsize=14)
plt.grid()
plt.show()
```

### 13.What is the relationship between adherence to treatment and treatment outcomes?

```python
adherence_by_outcome = (
    df.groupby('Outcome')['Adherence_to_Treatment_(%)']
    .mean()
    .reset_index()
    .rename(columns={'Adherence_to_Treatment_(%)': 'Avg Adherence (%)'})
)
print(adherence_by_outcome)
# Plotting the adherence to treatment by outcome
plt.figure(figsize=(10, 6))
plt.bar(adherence_by_outcome['Outcome'], adherence_by_outcome['Avg Adherence (%)'], color='skyblue')
plt.title('Relationship Between Adherence to Treatment and Treatment Outcomes', fontsize=16)
plt.xlabel('Treatment Outcome', fontsize=14)
plt.ylabel('Average Adherence (%)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 14.How does treatment duration correlate with stress levels?

```python
subset = df[['Treatment_Duration_(weeks)', 'Stress_Level_(1-10)']]

# Compute the correlation matrix
correlation_matrix = subset.corr()
print(correlation_matrix)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Heatmap: Correlation Between Treatment Duration and Stress Levels', fontsize=16)
plt.show()
```

### 15.What are the patterns in treatment start dates and their outcomes?

```python
monthly_outcomes = df.groupby(['Treatment_Start_Date', 'Outcome']).size().unstack()
print(monthly_outcomes)
# Plot distribution of start dates by month
plt.figure(figsize=(10, 6))
sns.countplot(x='Treatment_Start_Date', data=df, order=range(1, 13), palette="viridis")
plt.title('Distribution of Treatment Start Dates by Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Treatments', fontsize=14)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot outcome distribution by month
monthly_outcomes.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Treatment Outcomes by Start Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Treatments', fontsize=14)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], fontsize=12)
plt.legend(title='Outcome', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
